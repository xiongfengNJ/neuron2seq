```python
# tfidf

data=data_normal
x=data['region'].values
y=data['label_'].values
count_vect = CountVectorizer(analyzer="word",ngram_range=(1,1))
x_count = count_vect.fit_transform(x.copy())
    
tfidf_vect = TfidfVectorizer(analyzer='word',ngram_range=(1,1))
x = tfidf_vect.fit_transform(x)
x = x.toarray()

sss = StratifiedShuffleSplit(n_splits=2, test_size=0.2, random_state=0)
for train_index, test_index in sss.split(x, y):
    x_train=x[train_index]
    y_train=y[train_index]
    x_test=x[test_index]
    y_test=y[test_index]
```

```python
sentences=pd.DataFrame(x_count.toarray(),columns=list(tfidf_vect.vocabulary_.keys()),index=data.index)
sentences
```

```python
clf = OneVsRestClassifier(SGDClassifier(loss='hinge',max_iter=5000, tol=1e-3)).fit(x_train, y_train)
clf_y_pred = clf.predict(x_test)
(accuracy_score(y_test, clf_y_pred))
```

```python
clf = OneVsOneClassifier(SGDClassifier(loss='hinge',max_iter=5000, tol=1e-3)).fit(x_train, y_train)
clf_y_pred = clf.predict(x_test)
(accuracy_score(y_test, clf_y_pred))
```

```python
sgd=SGDClassifier(loss='hinge',max_iter=5000, tol=1e-3)
sgd.fit(x_train,y_train)
sgd_y_pred = sgd.predict(x_test)
accuracy_score(y_test, sgd_y_pred)
```

```python
tfidf_clf_acc_list=[]

for i in range(30):
    clf = OneVsOneClassifier(SGDClassifier(loss='hinge',max_iter=500, tol=1e-3)).fit(x_train, y_train)
    clf_y_pred = clf.predict(x_test)
    tfidf_clf_acc_list.append(accuracy_score(y_test, clf_y_pred))

figure_df = pd.DataFrame({'val_acc':tfidf_clf_acc_list, 'Methods':'TFIDF'})


path_han = '../result/final_training_result/HAN_region/from_1700/final_loss_valacc.csv'
han_df = pd.read_csv(path_han, sep=',',usecols=[1])
han_df['Methods']='HAN'
figure_df = pd.concat([figure_df,han_df],axis=0)

figure_df.reset_index(inplace=True, drop= True)

sns.set(font_scale=2)
plt.figure(figsize=(13,10))
sns.boxplot(data=figure_df,x='Methods',y='val_acc')
plt.xticks(rotation=10)
```
