```python
# pred_num: 代表该神经元属于哪一类--数字标志
# pred_num: numerical label for each class
data_ae['pred_num']=np.argmax(y_ae_class_pred,axis=1)
# print("data_ae['pred_num'] np.unique: ",np.unique(data_ae['pred_num']))
data_ae
```

```python
# trained type: 判断是否属于5类:不属于则为-1；属于为1；这是真实标签
# trainde_type: if this swc belongs to the 5 classes, set it to 1; else set it to -1; this is true label
data_ae['two_types_true']=-1
data_ae.loc[data_ae['label'].isin(['CP_GPe', 'CP_SNr', 'ET_MO', 'ET_SS', 'IT_MO', 'IT_SS', 'IT_VIS', 'LGd', 'MG',
 'RT', 'VPL', 'VPM']),'two_types_true']=1
data_ae
```

```python
# record the loss and score for each swc
new_scores=[]

for i in range(len(ae_pred_ae)):
    loss=mse(x_ae[i,:,:] , ae_pred_ae[i,:,:]).numpy()
    
    # transform the loss to score; 
    # higher score means swc is more likely to be classified to knonwn type.
    new_scores.append(1/abs(loss))

new_scores = np.array(new_scores)
new_scores=new_scores.reshape(len(new_scores),1)
new_scores
```

```python
# find the best threshold for the classifier
best_acc=0
best_th=-100
for th in new_scores:
    new_label_score = np.zeros((len(new_scores),1))-1
    new_label_score[new_scores>=th]=1
    cur_acc = metrics.accuracy_score(data_ae['two_types_true'],new_label_score)
    if cur_acc>=best_acc:
        best_acc=cur_acc
        best_th=th
print('best_acc',best_acc)
print('best thresh, ',best_th)

new_label_score = np.zeros((len(new_scores),1))-1
new_label_score[new_scores>=best_th]=1

metrics.accuracy_score(data_ae['two_types_true'],new_label_score)
```

```python
fpr, tpr, thresholds = metrics.roc_curve(data_ae['two_types_true'], new_scores, pos_label=1)
```

```python
plt.figure(figsize=(9,8))
print('auc: ',metrics.auc(fpr, tpr))

line=plt.plot(fpr,tpr,c='b',label='new cell detector: auc '+ str(round(metrics.auc(fpr, tpr),2)))
plt.xlabel('True positive rate',fontdict={'size' : 25})
plt.ylabel('False positive rate',fontdict={'size' : 25})
plt.tick_params(labelsize = 20)
plt.legend(handles=line,loc='lower right',fontsize=20)
# plt.savefig(r'ae_od.png',dpi=200)
```

```python
# if_trained_class : predicted label for unkonwn and known type
data_ae['if_trained_class']=new_label_score

# set pred num to -1 if the swc not belong to known type
data_ae.loc[data_ae.if_trained_class==-1,'pred_num']=-1

# pred region: real region name for each predicted numerical label
data_ae['pred_region'] = data_ae['pred_num'].map(lambda r: region_num_dict[r])
print("data_ae['pred_region'] np.unique: ",np.unique(data_ae['pred_region']))
```
