```python
from gensim.models import Word2Vec
from gensim.models import KeyedVectors
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import os
import pickle
import sys
sys.setrecursionlimit(1000000)
import pickle
import pandas as pd
import numpy as np

from keras.callbacks import LambdaCallback

from keras.utils.np_utils import to_categorical
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns

data_path = '../final_data/1728_seu_swc.pickle'
f = open(data_path, 'rb')
data = pickle.load(f)
f.close()
data = data[(data.label != 'unknown') & (data.label != 'CP_others')]
# df = pd.DataFrame(data['label'].value_counts()).astype(int)
# print('============ before ============')
# for k, v in data['label'].value_counts().items():
#     print(k, '=================', v)

# th = 30
# p = df.drop(df[df.label < th].index, axis=0).index.tolist()
# data = data[data.label.isin(p)]
# print('++++++++++++  after  ++++++++++++++++++')
print(data['label'].value_counts())


le = preprocessing.LabelEncoder()
y = le.fit_transform(data['label'])
data['label_'] = y
```

```python
# frequency of a region appears in the 405 cells dataset, to filter out regions with low frequency in word2vec training

x=data['region'].values
y=data['label_'].values
count_vect = CountVectorizer(analyzer="word",ngram_range=(1,1))
x_count = count_vect.fit_transform(x.copy())
sentences=pd.DataFrame(np.reshape(np.sum(x_count.toarray(),axis=0),(1,x_count.toarray().shape[1])),columns=list(count_vect.vocabulary_.keys()),index=[0])
region_in_dataset=pd.DataFrame({"region":list(count_vect.vocabulary_.keys()),'number':sentences.values.tolist()[0]})

plt.figure(figsize=(80,10))
plt.bar(region_in_dataset['region'],region_in_dataset['number'])
plt.ylim((0,150))
```

```python
print(np.percentile(region_in_dataset['number'],33))
```

```python
# for word2vec training, choose window size
num_continuous_region_list=[]
label_list=[]

for i in data.index.tolist():
    temp_nodes=data.loc[i,'region'].split(' ')
    num=1
    cur_region=temp_nodes[0]
    for j in temp_nodes[1:]:
        if j == cur_region:
            num=num+1
        else:
            cur_region = j
            num_continuous_region_list.append(num)
            label_list.append(data.loc[i,'label'])
            num=1
            

```

```python
num_continuous_region_df=pd.DataFrame({'Num-Continuous-Region':num_continuous_region_list,'label':label_list})
plt.figure(figsize=(10,10))
sns.kdeplot(data=num_continuous_region_df,x='Num-Continuous-Region')
plt.xlim((-30,300))
plt.savefig('Num-Continuous-Region.png',dpi=200)
```

```python
# After decomposing the neuron into several segmnets, to see the distribution of nodes number in one segment
sentence_length_list=[]
label_list=[]
for i in data.index.tolist():
    temp_nodes=list(data.loc[i,'node_type'])
    sentence_length=0
    for j in temp_nodes:
        sentence_length=sentence_length+1
        if j == 't':
            sentence_length_list.append(sentence_length)
            label_list.append(data.loc[i,'label'])
            sentence_length=0
```

```python
sentence_length_df=pd.DataFrame({'Segment-Length':sentence_length_list,'label':label_list})
sentence_length100_df=sentence_length_df[sentence_length_df.loc[:,'Segment-Length']<100]

sns.set(font_scale=2)
f,ax=plt.subplots(1,1,figsize=(10,10))
sns.kdeplot(data=sentence_length100_df,x='Segment-Length',ax=ax)
# sns.kdeplot(data=sentence_length_df,x='sentence_length',hue='label',ax=ax[1])
plt.savefig('.\segment_length.png',dpi=200)
```

```python
# 
num_segments_list=[]
label_list=[]
for i in data.index.tolist():
    temp_nodes=list(data.loc[i,'node_type'])
    num=0
    for j in temp_nodes:
        if j == 't':
            num=num+1
    num_segments_list.append(num)
    label_list.append(data.loc[i,'label'])
num_segments_df=pd.DataFrame({'Num-Segments':num_segments_list,'label':label_list})
```

```python
plt.figure(figsize=(10,10))
sns.kdeplot(data=num_segments_df,x='Num-Segments')
plt.savefig(r'number of segments.png',dpi=200)
```

```python
plt.figure(figsize=(12,10))
sns.kdeplot(data=num_segments_df,x='num_segments',hue='label')
# plt.savefig(r'number of sentencs.png',dpi=300)
```

```python
# length of neuron sequences
num_nodes=[]
label_list=[]
for i in data.index.tolist():
    temp_nodes=list(data.loc[i,'node_type'])
    num_nodes.append(len(temp_nodes))
    label_list.append(data.loc[i,'label'])
num_nodes_df=pd.DataFrame({'num_nodes':num_nodes,'label':label_list})
plt.figure(figsize=(10,10))
sns.kdeplot(data=num_nodes_df,x='num_nodes',hue='label')
```

```python
# when we set fixed length of sequence to 2000
prop_nodes=[]
label_list=[]
fixed_length=1900
for i in data.index.tolist():
    temp_nodes=list(data.loc[i,'node_type'])
    prop_nodes.append(len(temp_nodes)/fixed_length)
    label_list.append(data.loc[i,'label'])
prop_nodes_df=pd.DataFrame({'prop_nodes':prop_nodes,'label':label_list})
plt.figure(figsize=(10,10))
plt.axhline(y=1, color='y')
sns.boxplot(data=prop_nodes_df,y='prop_nodes')
```
